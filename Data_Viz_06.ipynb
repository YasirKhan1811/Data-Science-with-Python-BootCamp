# Count Plots

```python
# 1- imprort libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 2- import data from file
chilla = pd.read_excel("chilla_data_cleaned.xlsx", "Sheet1")

# 3- setting the Canvas
sns.set_theme(style="ticks", color_codes=True)

# 4- plot
p1 = sns.countplot(x = "Gender", data = chilla, 
                   palette="pastel", saturation=1)
plt.xlabel("Gender", size=14, weight="bold")
plt.ylabel("Count", size=14, weight="bold")
plt.title("Plot for Counting Chilla Participants", size=16, weight="bold")
```

*From this plot, it is concluded that the number of male participants in this chilla are up to 300 and the female participants are about 90.*

**The following countplot shows the numbers of male and female participants of this chilla based on their education completed.**

```python
# 1- imprort libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 2- import data from file
chilla = pd.read_excel("chilla_data_cleaned.xlsx", "Sheet1")

# 3- setting the Canvas
sns.set_theme(style="ticks", color_codes=True)

# 4- plot
p1 = sns.countplot(x = "Qualification_completed", hue="Gender",
                   data = chilla, palette="pastel", saturation=1,
                   order=["Pre-Bachelors", "Bachelors", "Masters", "PhD"])
p1.set_title("Plot for Counting Chilla Participants Based on Qualification")

# 5- setting and sizing labels and title
plt.xlabel("Qualification_completed", size=14, weight="bold")
plt.ylabel("Count", size=14, weight="bold")
plt.title("Participants Based on Qualification", size=16, weight="bold")
```

# Bar Plot

```python
# 1- imprort libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 2- import data from file
chilla = pd.read_excel("chilla_data_cleaned.xlsx", "Sheet1")

# 3- setting the Canvas
sns.set_theme(style="ticks", color_codes=True)

# 4- size of figure
plt.figure(figsize=(10,6))
 
# 5- plot
p1 = sns.barplot(x = "How many hours you code a day? (int) e.g: 5,4,3", 
                 y="field_of_study", 
                 data = chilla, palette="Set2", saturation=1,
                 linewidth=1, ci=None, edgecolor="0.1")

# 6- setting and sizing labels and title
plt.xlabel("Coding (Hrs)", size=14, weight="bold")
plt.ylabel("Field of Study", size=14, weight="bold")
plt.title("Hours of Coding Based on Different Education Backgrounds", size=16, weight="bold")
```

# Box Plots

```python
# 1- imprort libraries
import seaborn as sns
import pandas as pd

# 2- setting the Canvas
palette = {"Yes": "Grey", "No": "Purple"}

# 3- figure size
plt.figure(figsize=(10,6))

# 4- import data from file
chilla = pd.read_excel("chilla_data_cleaned.xlsx", "Sheet1")

# 5- Draw a boxplot
sns.boxplot(x="Research/Working experience (Float/Int) years", y="field_of_study", 
            palette="pastel",
            data=chilla)

# 6- setting axis limits and offset
plt.title("")
plt.xlim(-1,12)
sns.despine(offset=0, trim=False)

# 7- setting and sizing labels and title
plt.xlabel("Research/Working Experience", size=14, weight="bold")
plt.ylabel("Field of Study", size=14, weight="bold")
plt.title("Box Plot of Research Experience", size=16, weight="bold")
```

```python
# 1- import libraries

import seaborn as sns
import pandas as pd

# 2- setting theme
palette = {"Yes": "Yellow", "No": "Green"}

# 3- load the dataset
chilla = pd.read_excel("chilla_data_cleaned.xlsx", "Sheet1")

# 4- Draw a boxplot
sns.boxplot(x="How many hours you code a day? (int) e.g: 5,4,3", 
            y="What are you?", hue="Marital Status?", palette=palette,
            data=chilla, dodge=True)

# 5- offset and trim
sns.despine(offset=5, trim=False)

# 6- setting axis limits
plt.xlim(-1, 12)

# 7- setting and sizing labels and title
plt.xlabel("Research/Working Experience", size=14, weight="bold")
plt.ylabel("Field of Study", size=14, weight="bold")
plt.title("Daily Coding Hours as per the Employment Status", size=16, weight="bold")
```
