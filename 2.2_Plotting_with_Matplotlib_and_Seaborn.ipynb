**Adding Titles**

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
# phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")

plt.show()
```

**Adding Limits**

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
# phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(4)
plt.ylim(2.2)

plt.show()
```

**Set Styles**

- darkgrid
- whitegrid
- dark
- white
- ticks

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt


# load dataset
phool = sns.load_dataset("iris")
# phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
sns.set_style(style=None, rc=None)
sns.set_style("darkgrid")

plt.show()
```

### Figure Size

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
# phool

# size of figure
plt.figure(figsize=(10,6))

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")

plt.show()
```
