# Pandas Case Study

#### This dataset is Cereals Production data from the year 1961 to 2019, and it was taken from FAOSTAT (Food and Agriculture Organization of the United Nations). In the following step by step procesdure, I am going to use Pandas and NumPy libraries to perform data anaylsis and data manipulation on this dataset. 

First step is to import the following libraries:

```python
import pandas as pd
import numpy as np
```

Reading Data from a CSV file

```python
usa_cereals = pd.read_csv("FAOSTAT_data_USA.csv")
usa_cereals.head(3)
```

Showing basic statistics and summary

```python
usa_cereals.describe()
```

Dropping few columns and making new dataset and assigning the new dataset to a variable "a"

```python
a = usa_cereals.drop(["Domain Code", "Element Code", "Year Code", "Domain", "Area Code", "Item Code", "Unit"], axis=1)
```

```python
a
```

Finding the mean production per year associated with the element category.

```python
a.groupby(["Element", "Year"]).mean()
```

```python
a[a["Value"] > 6.076251e+07].groupby(["Element", "Year"]).mean()
```
